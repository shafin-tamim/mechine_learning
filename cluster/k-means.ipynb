```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('F:\mechine_learning\Data\Iris - Copy.csv')
```

```python
df.head(2)
```

```python
plt.figure(figsize=(5, 5))
sns.pairplot(data=df)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(2, 20):
    kmeans=KMeans(n_clusters=i, init='k-means++')
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10, 5))
plt.plot([i for i in range(2, 20)], wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.xticks([i for i in range(2, 20)])
plt.ylabel('WCSS')  
plt.grid(axis="x")
plt.show()     
```

```python
knn=KMeans(n_clusters=3, init='k-means++')
df['pradict']=knn.fit_predict(df)
```

```python
df.head(5)
```

```python
knn.labels_
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(df,labels=knn.labels_)
```

```python
sns.pairplot(df, hue='pradict')
plt.show
```

```python
ss=[]
no_c=[j for j in range(2, 20)]
for i in range(2, 20):
    knn1=KMeans(n_clusters=i)
    knn1.fit(df)
    ss.append(silhouette_score(df, labels=knn1.labels_))
```

```python
plt.plot(no_c, ss, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.xticks(no_c)
plt.grid(axis="x")
plt.title('Silhouette Score vs Number of clusters')
plt.show
```
