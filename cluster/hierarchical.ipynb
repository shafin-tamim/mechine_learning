```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('F:\mechine_learning\Data\Iris - Copy.csv')
```

```python
df.head(2)
```

```python
plt.figure(figsize=(5, 5))
sns.pairplot(data=df)
plt.show()
```

```python
import scipy.cluster.hierarchy as sch
```

```python
plt.figure(figsize=(10, 5))
sch.dendrogram(sch.linkage(df, method='single', metric='euclidean'))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac=AgglomerativeClustering(n_clusters=2, linkage='single')
df['pradict']=ac.fit_predict(df)
```

```python
sns.pairplot(data=df, hue='pradict')
plt.show()
```
